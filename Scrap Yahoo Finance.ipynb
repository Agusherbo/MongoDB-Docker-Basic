### Extraer datos de yahooFinance

```python
import yfinance as yf # libreria para obtener datos de Yahoo Finance
import pandas as pd
import numpy as np
from datetime import datetime
import json
```

```python
#Definimos los tickers de las empresas de las que queremos sacr informaci√≥n
tickers = ['TSLA', 'NFLX', 'META', 'AMZN', 'GOOG']
```

```python
tic = yf.Ticker(tickers[0])
tic.info
```

```python
#Localizamos el indice de los elementos que queremos
for i,e in enumerate(tic.info.items()):
    print(i,e)
```

```python
# Localizamos los elementos que queremos en una informacion anidada
for i,e in enumerate(tic.info['companyOfficers'][0].items()):
    print(i,e)
    
```

```python
#Localizamos los nombre de la info que queremos extraer
tic_fin = tic.financials
tic_fin
```

```python
#Codigo para generar el documento
jason = []

for i in range(len(tickers)):
    d = {}
    tick = yf.Ticker(tickers[i])
    for ix,e in enumerate(tick.info['companyOfficers'][0].items()):
        if ix == 1 or ix == 3 or ix == 4:
            k, v = e[0], e[1]
            d[k] = v
    
    for ix,e in enumerate(tick.info.items()):
        if ix == 7 or ix == 13:
            k, v = e[0], e[1]
            d[k] = v
    
    t_fin = tick.financials
    for ix,e in enumerate(t_fin.index):
        if e == 'Gross Profit' or e == 'Cost Of Revenue' or e == 'Total Revenue':
            k, v = e, t_fin.iloc[ix][0]
            d[k] = v
   
    jason.append(d)

```
